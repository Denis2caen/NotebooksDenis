Dans la cellule ci-dessous, écrire les instructions permettant d'obtenir les listes suivantes :
 - $\texttt{LAT}$ contenant les latitudes
 - $\texttt{LONG}$ contenant les longitudes
 - $\texttt{DONNEES}$ contenant les fréquentations

```python
LAT =  list(monFichier['latitude'])  
print(LAT)
LONG =   list(monFichier['longitude'])  
print(LONG)
DONNEES =   list(monFichier['Nombre_d_eleves'])  
print(DONNEES)
```

```python
LAT = list(monFichier['latitude'])
LONG =  list(monFichier['longitude'])
DONNEES = list(monFichier['Nombre_d_eleves'])
```

On aura besoin des coordonnées GPS sous forme d'un couple (latitude, longitude).  
Pour cela, on va construire la liste de tous les couples de coordonnées en écrivant les instructions suivantes :

```python
GPS = [] 
for i in range(len(LAT)):  
    coordGPS = (LAT[i],LONG[i])
    GPS.append(coordGPS) 
```

* Pour centrer notre carte, je vous propose de choisir le point dont la latitude est la moyenne entre la plus petite et la plus grande latitude et la longitude est  la moyenne entre la plus petite et la plus grande longitude. 

```python
lat_centre = (min(LAT)+max(LAT))/2
long_centre = (min(LONG)+max(LONG))/2
GPS_centre = (lat_centre,long_centre)
```

* On import le module folium dans une cellule pour pouvoir construire la carte et y placer les marqueurs

```python
import folium
```

*  On onstruit la carte qui contiendra les effectifs en choisissant le centre et le zoom

```python
carte  = folium.Map(location = GPS_centre,  zoom_start = 8)
carte
```

* On va ajouter à cette carte des disques dont l'aire est proportionnelle au nombre de festivaliers qu'il représente autrement dit le rayon est proportionnel à la racine carrée du nombre de festivaliers. Pour que le disque ne soit pas trop grand, il est nécessaire de bien choisir le coefficient de proportionnalité. Pour cela regardons les racines carrées du plus petit et du plus grand nombre.

La fonction racine carrée doit être importée du module math pour pouvoir être utilisée

```python
from math import sqrt
```

```python
print(sqrt(min(DONNEES)), sqrt(max(DONNEES)))
```

 Il faut parfois faire plusieurs essais pour obtenir ce qu'on veut : je propose ici de prendre un coefficient de 1/50. 

```python
for i in range(len(NOMS)):
    message = NOMS[i] +' : '+ str(DONNEES[i])
    rayon =  sqrt(DONNEES[i])/5
    folium.CircleMarker(location = GPS[i], radius = rayon,color = 'crimson',fill = True,fill_color = 'crimson'
                        ,tooltip = message).add_to(carte)

carte
```
